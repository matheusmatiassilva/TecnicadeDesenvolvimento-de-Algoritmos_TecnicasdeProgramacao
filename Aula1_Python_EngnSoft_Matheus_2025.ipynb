```python
def verificar_multiplo_seguro():
    """Verifica a multiplicidade, incluindo tratamento de erros para entrada e divisão por zero."""

    # 1. Leitura e Validação do Número (deve ser um inteiro)
    while True:
        try:
            numero = int(input("Digite um número inteiro: "))
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

    # 2. Leitura e Validação do Múltiplo (deve ser um inteiro e não pode ser 0)
    while True:
        try:
            multiplo = int(input("Digite o múltiplo a ser testado (diferente de 0): "))
            if multiplo == 0:
                print("O múltiplo não pode ser zero. Por favor, digite outro valor.")
            else:
                break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

    # 3. Lógica de Verificação (igual ao seu código)
    if numero % multiplo == 0:
        print(f"\nResultado: O número {numero} é múltiplo de {multiplo}.")
    else:
        print(f"\nResultado: O número {numero} não é múltiplo de {multiplo}.")

# Para testar:
# verificar_multiplo_seguro()
```

```python
#9
def q09_hello_world_n_vezes():
    """Solicita N e imprime 'Hello World!' N vezes, uma por linha."""

    while True:
        try:
            # Solicita a quantidade de repetições
            n = int(input("Digite um número inteiro (N) para a quantidade de repetições: "))

            if n >= 0:
                break # Sai do loop se a entrada for um inteiro não negativo
            else:
                print("Por favor, digite um número inteiro não negativo.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

    print(f"\nImprimindo 'Hello World!' {n} vezes:")

    # Loop para imprimir N vezes
    for _ in range(n):
        print("Hello World!")

# q09_hello_world_n_vezes()
```

```python
#11
def q11_soma_e_media_n():
    """Lê N, depois lê N números e calcula a soma e a média."""

    while True:
        try:
            # 1. Solicita a quantidade de números (N)
            n = int(input("Digite a quantidade de números (N) a serem lidos: "))

            if n >= 0:
                break
            else:
                print("N deve ser um número inteiro não negativo.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

    if n == 0:
        print("Nenhuma leitura será feita. Soma = 0, Média = 0.")
        return

    soma = 0

    # 2. Loop para ler N números e calcular a soma
    for i in range(n):
        while True:
            try:
                # Solicita o número e adiciona à soma
                numero = float(input(f"Digite o {i + 1}º número: "))
                soma += numero
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número.")

    # 3. Calcula a média
    media_aritmetica = soma / n

    # 4. Apresenta os resultados
    print("\n--- Resultados ---")
    print(f"A Soma dos {n} números é: {soma:.2f}")
    print(f"A Média Aritmética é: {media_aritmetica:.2f}")

# q11_soma_e_media_n()
```

```python
#12
def q12_pessoa_mais_velha():
    """Lê N, depois lê nome e idade de N pessoas e encontra a mais velha."""

    while True:
        try:
            # 1. Solicita a quantidade de registros (N)
            n = int(input("Digite a quantidade de pessoas (N) a serem registradas: "))
            if n >= 0:
                break
            else:
                print("N deve ser um número inteiro não negativo.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

    if n == 0:
        print("Nenhum registro será feito.")
        return

    # Inicializa as variáveis para rastrear a pessoa mais velha
    mais_velha_nome = ""
    mais_velha_idade = -1 # Um valor que será imediatamente substituído

    # 2. Loop para ler Nome e Idade de N pessoas
    for i in range(n):
        print(f"\n--- Registro {i + 1} de {n} ---")
        nome = input("Digite o nome da pessoa: ")

        while True:
            try:
                idade = int(input("Digite a idade da pessoa: "))
                if idade >= 0:
                    break
                else:
                    print("A idade deve ser um número inteiro não negativo.")
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro para a idade.")

        # 3. Verifica se a pessoa atual é a mais velha
        if idade > mais_velha_idade:
            mais_velha_idade = idade
            mais_velha_nome = nome

    # 4. Apresenta o resultado
    print("\n--- Resultado Final ---")
    print(f"A pessoa mais velha é: {mais_velha_nome}")
    print(f"Com a idade de: {mais_velha_idade} anos")

# q12_pessoa_mais_velha()
```

```python
#13
def q13_media_idade_genero():
    """Lê N, depois lê sexo e idade de N pessoas e calcula a média de idade por gênero."""

    while True:
        try:
            # 1. Solicita a quantidade de registros (N)
            n = int(input("Digite a quantidade de pessoas (N) a serem registradas: "))
            if n >= 0:
                break
            else:
                print("N deve ser um número inteiro não negativo.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

    if n == 0:
        print("Nenhum registro será feito.")
        return

    # Inicializa as variáveis para acumular a soma das idades e a contagem
    soma_idade_masc = 0
    count_masc = 0
    soma_idade_fem = 0
    count_fem = 0

    # 2. Loop para ler Sexo e Idade de N pessoas
    for i in range(n):
        print(f"\n--- Registro {i + 1} de {n} ---")

        # Leitura e validação do sexo
        while True:
            sexo = input("Digite o sexo (M/F): ").upper().strip()
            if sexo in ('M', 'F'):
                break
            else:
                print("Entrada inválida. Digite apenas M ou F.")

        # Leitura e validação da idade
        while True:
            try:
                idade = int(input("Digite a idade da pessoa: "))
                if idade >= 0:
                    break
                else:
                    print("A idade deve ser um número inteiro não negativo.")
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro para a idade.")

        # 3. Acumula a soma e a contagem de acordo com o gênero
        if sexo == 'M':
            soma_idade_masc += idade
            count_masc += 1
        elif sexo == 'F':
            soma_idade_fem += idade
            count_fem += 1

    # 4. Calcula e apresenta as médias

    # Média Masculina
    if count_masc > 0:
        media_masc = soma_idade_masc / count_masc
        print(f"\nMédia de Idade Masculina ({count_masc} registros): {media_masc:.2f} anos")
    else:
        print("\nNenhum registro masculino encontrado.")

    # Média Feminina
    if count_fem > 0:
        media_fem = soma_idade_fem / count_fem
        print(f"Média de Idade Feminina ({count_fem} registros): {media_fem:.2f} anos")
    else:
        print("Nenhum registro feminino encontrado.")

# q13_media_idade_genero()
```

```python
#14
def q14_maior_e_menor_10_numeros():
    """Solicita 10 números reais e imprime o maior e o menor."""
    print("--- Exercício 14: Maior e Menor de 10 Números ---")

    numeros = []

    for i in range(10):
        while True:
            try:
                numero = float(input(f"Digite o {i + 1}º número real: "))
                numeros.append(numero)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número real.")

    if numeros:
        maior = max(numeros)
        menor = min(numeros)

        print("\n--- Resultados ---")
        print(f"O maior número informado é: {maior}")
        print(f"O menor número informado é: {menor}")
    else:
        print("Nenhum número foi inserido.")

# q14_maior_e_menor_10_numeros()
```

```python
#15
def q15_ordenar_numeros_ate_zero():
    """Solicita números reais até que zero seja informado, então ordena e exibe a lista."""
    print("--- Exercício 15: Ordenação de Números (Parada no Zero) ---")

    numeros = []

    while True:
        while True:
            try:
                entrada = float(input("Digite um número real (ou 0 para ordenar e finalizar): "))
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número real.")

        if entrada == 0:
            break
        else:
            numeros.append(entrada)

    if not numeros:
        print("\nNenhum número foi informado para ordenação.")
        return

    numeros.sort() # Ordena a lista em ordem crescente

    print("\n--- Resultados ---")
    print("Os números informados, em ordem crescente, são:")
    for num in numeros:
        print(f"-> {num}")

# q15_ordenar_numeros_ate_zero()
```

```python
#16
def q16_analise_idades_alunos():
    """Coleta idades e calcula estatísticas: maior, menor, contagens, média e mediana."""
    print("--- Exercício 16: Análise Estatística de Idades ---")

    idades = []

    while True:
        try:
            # Pergunta se deseja informar a idade do próximo aluno
            continuar = input("Deseja informar a idade do próximo aluno? (S/N): ").upper().strip()

            if continuar == 'N':
                break # Sai do loop principal
            elif continuar != 'S':
                print("Opção inválida. Digite S para continuar ou N para finalizar.")
                continue # Volta ao início do loop

            # Solicitar e validar a idade
            while True:
                try:
                    idade = int(input("Digite a idade do aluno: "))
                    if idade >= 0:
                        idades.append(idade)
                        break
                    else:
                        print("A idade deve ser um número inteiro não negativo.")
                except ValueError:
                    print("Entrada inválida. Por favor, digite um número inteiro para a idade.")

        except EOFError: # Trata o caso de final de arquivo em alguns ambientes
            break


    if not idades:
        print("\nNenhuma idade foi informada. Análise não pode ser realizada.")
        return

    print("\n--- Resultados da Análise ---")

    # 1. Idade do mais novo e mais velho
    idade_mais_novo = min(idades)
    idade_mais_velho = max(idades)

    print(f"Total de Alunos: {len(idades)}")
    print(f"Idade do Aluno Mais Novo: {idade_mais_novo}")
    print(f"Idade do Aluno Mais Velho: {idade_mais_velho}")

    # 2. Contagens
    alunos_mais_de_18 = sum(1 for idade in idades if idade > 18)
    alunos_ate_18 = len(idades) - alunos_mais_de_18 # O restante tem <= 18 anos

    print(f"Alunos com MAIS de 18 anos (> 18): {alunos_mais_de_18}")
    print(f"Alunos com ATÉ 18 anos (<= 18): {alunos_ate_18}")

    # 3. Média Aritmética
    soma_idades = sum(idades)
    media_aritmetica = soma_idades / len(idades)
    print(f"Média Aritmética das Idades: {media_aritmetica:.2f}")

    # 4. Mediana
    idades_ordenadas = sorted(idades)
    n = len(idades_ordenadas)

    if n % 2 == 1:
        # Ímpar: valor central
        mediana = idades_ordenadas[n // 2]
    else:
        # Par: média dos dois centrais
        indice_superior = n // 2
        indice_inferior = indice_superior - 1
        mediana = (idades_ordenadas[indice_inferior] + idades_ordenadas[indice_superior]) / 2

    print(f"Mediana das Idades: {mediana:.2f}")

# q16_analise_idades_alunos()
```
